```python
current_df = pd.read_csv("Data/DFTrainTest.csv")
current_df
```

```python
camera_df = pd.read_csv("Data/Police_Crime_Cameras.csv")
camera_df.dropna(axis=0, inplace=True)
camera_df
```

```python
from math import sin, cos, sqrt, atan2, radians

# approximate radius of earth in km
def distanceBetweenPoints(lat1, lon1, lat2, lon2):
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c

    return distance
```

```python
for index, row in current_df.iterrows():
    for index2, row2 in camera_df.iterrows():
        current_df.set_value(index, 'Cam_Present', 0)
        if distanceBetweenPoints(row2['LATITUDE'], row2['LONGITUDE'], row['Latitude (X)'], row['Longitude (Y)']) < 1:
            current_df.set_value(index, 'Cam_Present', 1)
            break
        
            
            
```
